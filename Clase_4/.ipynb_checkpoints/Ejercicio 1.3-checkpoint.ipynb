# Ejercicio  1.3

Generar 100 obs del modelo:

$$x_t = -0.9 x_{t-2} +w_t$$

Con $\sigma_w = 1$, posteriormente aplicar el siguiente filtro de media m√≥vil:
$$v_t = (x_t+x_{t-1}+x_{t-2}+x_{t-3})/4$$

```python
#Generamos un vector con n = 100 + 2 realizaciones de una variable aleatoria con media 0 y varianza 1.
import numpy as np
import matplotlib.pyplot as plt
w = np.random.randn(100)

#Generamos las realizaciones del vector autoregresivo
x = np.zeros(102)
x[0] = 0
x[1] = 0
for i in range(2,len(w)):
    x[i]= -0.99 * x[i-2]+ w[i]

#Graficamos los resultados
plt.plot(x)
```

```python
#Aplicamos el filtro deseado
filtro = np.zeros(100)
filtro[0] = 0
filtro[1] = 0
filtro[2] = 0
filtro[3] = 0
for i in range(4,len(filtro)):
    filtro[i] = 0.25*(x[i]+x[i-1]+x[i-2]+x[i-3])
    
plt.plot(filtro)
    
```
