```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tickers = ['AAPL', 'MSFT', '^GSPC']

# We would like all available data from 01/01/2000 until 12/31/2016.
start_date = '2010-01-01'
end_date = '2020-12-31'

# User pandas_reader.data.DataReader to load the desired data. As simple as that.
panel_data = data.DataReader(tickers, 'yahoo', start_date, end_date)
```

```python
#Shift +tab en una función
panel_data['Adj Close'].plot(figsize=(15,15),grid = True,title = "Series de precios");
```

# Análisis descriptivo de las series históricas

```python
#Análisis expl
panel_data.describe()
```

```python
#diff = p(t)-p(t-1)
absolute_returns =panel_data.diff()
```

```python
panel_data.diff(periods = 2)
```

```python
absolute_returns['Adj Close'].plot(kind= 'box',figsize=(20,20),legend = False,grid = True)
```

```python
#pct_change (P(t)-P(t-1))/P(t-1)
relative_returns = panel_data.pct_change()
```

```python
relative_returns['Adj Close'].plot(kind= 'box',figsize=(20,20),legend = False,grid = True)
```

```python
rendimientos_logaritmicos = np.log(panel_data).diff()
```

```python
rendimientos_logaritmicos['Adj Close'].plot(kind= 'box',figsize=(20,20),legend = False,grid = True)
```
